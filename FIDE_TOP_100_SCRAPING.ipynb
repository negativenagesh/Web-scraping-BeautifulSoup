```python
with open('/FIDE_Ratings_and_Statistics.html') as fide:
  sp1=BeautifulSoup(fide,'lxml')
```

```python
for link in sp.find_all('a'):
  print(link.get('href'))
```

```python
list1=sp1.find_all('td',class_='flag-wrapper')
list1
```

```python
for name in sp1.find_all('a'):
  print(name.text)
```

```python
list2=sp1.find_all('td',class_='flag-wrapper')
print(list2)
```

```python
Country=[]
for item in list2:
    country_name = item.text.strip().split()[-1]
    Country.append(country_name)
Country
```

```python
ratings=[]
for row in sp1.find_all('tr')[1:]:
  rating=row.find_all('td')[3].text
  ratings.append(rating)
ratings
```

```python
names=[]
for row in sp1.find_all('tr')[1:]:
  name=row.find_all('td')[1].text
  names.append(name)
names
```

```python
Birth_year=[]
for row in sp1.find_all('tr')[1:]:
  byear=row.find_all('td')[5].text
  Birth_year.append(byear)
Birth_year
```

```python
avg12m=[]
for row in sp1.find_all('tr')[1:]:
  avg12=row.find_all('td')[-1].text
  avg12m.append(avg12)
avg12m
```

```python
import pandas as pd
data=pd.DataFrame()
```

```python
data['Name']=names
data['Rating']=ratings
data['Country']=Country
data['Birth year']=Birth_year
data
```

```python
data['Rating']=data['Rating'].astype(int)
data['Birth year']=data['Birth year'].astype(int)
```
